```python
import pandas as pd
import numpy as np

df= pd.read_csv('deliveries.csv')
df.head(1)
```

```python
df1=df[["match_id","batting_team","bowling_team","over","ball","total_runs","is_wicket","player_dismissed"]]
df1.head(3)
```

```python
df1.isnull().sum()
```

```python
df1["player_dismissed"] = df1["player_dismissed"].fillna("0")

df1.head(3)
```

```python
df1["total_runs"] = pd.to_numeric(df1["total_runs"], errors="coerce")

df1["cum_runs"] = df1.groupby("match_id")["total_runs"].cumsum()
df1.head(3)
```

```python
df1["balles_balled"]=(df1["over"].astype("int")*6)+df1["ball"].astype("int")
df1
```

```python
df1["ball_left"]=120-df1["balles_balled"]
df1["ball_left"]=df1["ball_left"].apply(lambda x:0 if x<0 else x)
df1.head(4)
```

```python
# Step 1: Make sure 'is_wicket' is binary (0 or 1)
df1["is_wicket"] = df1["is_wicket"].apply(lambda x: 1 if str(x) != "0" and str(x) != "NaN" else 0)

# Step 2: Cumulative wickets fallen per match (resets per match_id)
df1["wickets_fallen"] = df1.groupby("match_id")["is_wicket"].cumsum()

# Step 3: Wickets left = 10 - fallen wickets (reset for each match)
df1["wickets_left"] = 10 - df1["wickets_fallen"]

# Step 4: Ensure wickets_left never goes below 0
df1["wickets_left"] = df1["wickets_left"].clip(lower=0)

df1.head(500)
```

```python
df1["crr"]=(df1["cum_runs"]*6)/df1["balles_balled"]
df1
```

```python
# Ensure total_runs is numeric
df1["total_runs"] = pd.to_numeric(df1["total_runs"], errors="coerce")

# Group by match_id
groups = df1.groupby("match_id")
match_ids = df1["match_id"].unique()

last_five = []

for match_id in match_ids:
    match_df = groups.get_group(match_id).copy()
    # Rolling sum with window=30 (first 30 values will be NaN)
    match_df["last5_runs"] = match_df["total_runs"].rolling(window=30).sum()
    last_five.extend(match_df["last5_runs"].tolist())

df1["last5_runs"] = last_five  # attach back to main df if needed


```

```python
# last_five
```

```python
# Runs per match per batting team (acts as innings)
inning_runs = (
    df1.groupby(["match_id", "batting_team"])["total_runs"]
    .sum()
    .reset_index()
    .rename(columns={"total_runs": "inning_runs"})
)

# Merge back into df1 so each row knows its team's total runs
df1 = df1.merge(inning_runs, on=["match_id", "batting_team"], how="left")

print(df1.head())
```

```python
df2=df1[["batting_team","bowling_team","cum_runs","ball_left","wickets_left","crr","last5_runs","inning_runs"]]
df2
```

```python
df1.dropna(inplace=True)
```

```python
df2=df2.sample(df2.shape[0])
df2
```

```python
x=df2.drop(columns=["inning_runs"])
y=df2["inning_runs"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
# x_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(drop = "first", sparse_output=False),["batting_team","bowling_team"])
],remainder= 'passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler(with_mean=False)),
    ('step3',XGBRegressor(n_estimators = 1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
print(r2_score(y_test,y_pred))
```
